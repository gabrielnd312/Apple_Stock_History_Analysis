Apple is one of the largest companies in the world in the field of electronics and software. It was founded in 1976 by **Steve Jobs, Steve Wozniak and Ronald Wayne**.

The duo attempted to sell their project to **HP** and **Atari**, but both companies declined, stating that **personal computers had no future in the market**. Faced with the disbelief of major companies in their work, the pair decided to produce computers using their own resources and sell them independently.

This analysis aims to highlight key milestones in the company's history through its stock value, indicating significant market moments, the stability of its growth, and its current situation.

#**Dictionary of Variables**


* **Date** -> yyyy/mm/dd
* **Open** -> The share price at market opening
* **High** -> The maximum value reached by the share price in that period
* **Low** -> The minimum value reached by the share price in that period
* **Close** -> The share price at the close of that period
* **Adj Close** -> Adjusted close - Closing price after adjustments for all applicable splits and dividend distributions.
* **Volume** -> Shows the number of shares traded in the period

#**Graphic Visualization**

```python
df.hist(bins = 10, figsize = (15,10), color = "green");

plt.tight_layout()
```

Looking at these histograms it is easy to identify this long period of **low values.**

When we check the most recent entries we see the volume increase a little more.

Let's take a closer look at the "**Close**" graph

```python
df.hist(column=["Close"], bins=10, figsize=(10, 6), color="purple")

plt.grid(False)
plt.tight_layout()
```

With this augmented view we can visually identify what our `.describe()` method returned.

Most of the time the shares were below **US 25.00** with greater volatility from US 50.00 until the volume starts to return from **US 125.00**.

In order to verify what happened in this second interval, we will only visualize with the **information from the value of US 100.00**.

```python
df_close_100 = df[df["Close"]> 100]["Close"]
df_close_100.hist(bins = 10, figsize = (10,6), color = "#452256")

plt.grid(False)

plt.tight_layout()
```

With a view filtered from **100 dollars**, the highest trading volume is identified in the ranges from **130 to 140 dollars**, and then **150 dollars**.

But in general, we can easily see that the dispersion is much smaller and the **growth** of the market value since its IPO is very significant.


We can also verify this effect in a line graph. See the difference between them.

```python
fig, ax = plt.subplots (figsize = (10,6))

df_date_100 = df[df["Close"]> 100]["Date"]

ax.plot(df["Date"], df["Close"])
ax.set_title("All Close Values")
ax.set_xlabel("Date")
ax.set_ylabel("Share Price")

plt.xticks(rotation=90)

num_ticks = 20
x_ticks_index = range(0, len(df["Date"]), len(df["Date"]) // num_ticks)
x_ticks_labels = df["Date"][x_ticks_index]
ax.set_xticks(x_ticks_index)

plt.tight_layout()
```

#**Top 10 values**

<left><img alt="apple_stock9" width="60%" src="https://images.unsplash.com/flagged/photo-1578928534298-9747fc52ec97?ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8fA%3D%3D&auto=format&fit=crop&w=1332&q=80"></left>



To better understand our stock's current pricing and its position within history, let's look at the **highest quoted values** ​​for this stock.

```python
df_max = df[df["Close"]> 160]
df_max.sort_values(by = "Close", ascending=False)
```

The table above indicates, ordered by the value in the "Close" column, the **highest to the lowest value** ever found for the stock within the **10 highest values**.

We can verify, as previously indicated, that the highest historical value with the data present in this DataFrame is **188** dollars.

The second highest value is around **177** dollars and the last value on this list, the tenth highest value, is **162** dollars.

Well, with the information we have so far, including the mean, maximum values, median and also graphical views, we can understand that although there is growth, it is not **absolutely linear**.

What makes this conclusion easier is the analysis of the previous charts.

However, let's check within this list of the **10 largest values**, the dates on which these occurrences existed and check the difference between these dates to better understand the current moment of the action in the market.

```python
df_max_sorted = df_max.sort_values(by = "Close", ascending=False)
dates_max_value = df_max_sorted["Date"]

print(dates_max_value)

```

Among the **top 10 closing values**, we find:

* `4` in 2023
* `4` in 2022
* `2` in 2021


In the case of a historical analysis, we can therefore conclude that in recent years Apple has experienced its **best moment** as a company and in the midst of the market.



```python
#Bar Chart
plt.bar(df_max["Date"],df_max["Close"], color = "#245c70")

plt.xlabel("Date")
plt.ylabel("Share Price (US $)")
plt.title("Top 10 Share Price Over Time")

for i, value in enumerate(round(df_max["Close"])):
    plt.text(i, value, str(value), ha='center', va='bottom')

plt.xticks(rotation = 90)

plt.show()
```

It is then identified that indeed in 2023 the share prices continue at an increasing pace, however it is interesting to see that even in recent years the growth **does not occur uniformly**, with the second highest share value historically being in `December 2021`, and after that, the value was only higher in `June 2023`.

In that time, the stock rose to values ​​close to:

* `$ 177` in May 2023

* `$ 177` in April 2023

* `$ 175` in March 2022

* `$ 175` in January 2022




```python
#Bar Chart
plt.bar(df_max["Date"],df_max["Volume"], color = "#245c70")

plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Top 10 Share Price - Volume Over Time")


plt.xticks(rotation = 90)

plt.show()
```

#**Trading Volume Over Time**

```python
plt.bar(df["Date"],df["Volume"], color = "#245c70", width = 2)

plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Trading volume over time")

num_ticks = 10
x_ticks_index = range(0, len(labels_all), len(labels_all) // num_ticks)
x_ticks_labels = labels_all[x_ticks_index]
plt.xticks(x_ticks_index, x_ticks_labels, rotation=90)

plt.xticks(rotation = 90)

plt.show()
```

**Volume analysis** shows us that despite the company's value continuing to grow through the years, the trading volume grew to its peak around 2008 - 2009, and then continued to decline for several years until the present day.

A deeper analysis of the macroeconomic context is needed, but this brings up possible reflections:

Greater confidence in the company, which causes a lower volume of daily sales, reducing the trading volume. Feeling of economic insecurity on the part of investors, less willing to invest recurrently and more willing to insure their investments already made.

Based on price and volume analysis, we can take new parameters to build other scenarios.
